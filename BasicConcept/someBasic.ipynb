```python
"""
AUC是ROC曲线下面积，在0到1之间，为0，说明正负样本分反，0.5为随机猜测，1为完美预测。
横坐标是：False Positive Rate，即FP/(FP+TN)，表示模型预测为正的样本中，实际为负的比例。
纵坐标是：True Positive Rate，即TP/(TP+FN)，表示模型预测为正的样本中，实际为正的比例。
"""

from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# 示例数据（真实标签和预测概率）
y_true = [0, 1, 1, 0, 1, 1, 0, 0]
y_scores = [0.1, 0.9, 0.8, 0.2, 0.7, 0.6, 0.3, 0.4]

# 计算FPR, TPR, 阈值
fpr, tpr, thresholds = roc_curve(y_true, y_scores)

# 计算AUC
auc = roc_auc_score(y_true, y_scores)
print("AUC:", auc)  # 输出: 0.9375

# 绘制ROC曲线
plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.plot([0, 1], [0, 1], "k--")  # 随机猜测线
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.show()
```
